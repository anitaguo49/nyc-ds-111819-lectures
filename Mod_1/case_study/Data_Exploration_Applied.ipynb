## 1. Handle the non-normal values

As you can see from the data dictionary, there are a number of responses that are coded with values that could throw off our descriptive statistics.  Let's clean up by mapping the responses to a more appropriate value, and handling any missing values.  

```python
#your code here
#1.create dic with value mappings
#2.use dic with the replace method
health_replace={88:0,
               77:np.nan,
               99:np.nan}
df['PHYSHLTH'].replace(health_replace, inplace=True) #save the changes in actual(orig) dataset; syntax-> string.replace(old,new,count)
```

```python
df.shape
```

```python
#drop the nans, add codec.
df.dropna(subset=['PHYSHLTH'],inplace=True)
```

## 3. Generate a graphic that compares the average number of sick days for each state. 

```python
# code to calculate the average number of sick days by state
avg = df.groupby('_STATE')['PHYSHLTH'].mean()
avg
```

```python
states = ['Connecticut','NJ','NY']
means = list(avg.values)
```

```python
# code to graph those averages using matplotlib
fig,ax = plt.subplots()

ax.bar(states,means,width=0.8,align='center')
ax.set_xticklabels(states)
ax.set_ylabel('Average per State')
ax.set_xlabel('States')

plt.show()
```

```python
# graph a histogram
fig,ax = plt.subplots()

ax.hist(df[df['_STATE']==9]['PHYSHLTH'],histtype='step',label='Conn')
ax.hist(df[df['_STATE']==34]['PHYSHLTH'],histtype='step',label='NJ')
ax.hist(df[df['_STATE']==36]['PHYSHLTH'],histtype='step',label='NY')

ax.set_ylabel('Frequencies')
ax.set_xlabel('States')
ax.set_title('Compare Sick Days by States')
ax.legend()

plt.show()
```

## 4. Turn the number of sick days into a classification of of chronically sick people.  

Chronically sick people account for a significant proportion of the costs for health insurers. If we can identify what leads people to be chronically sick, we can address them with specific treatments.



```python
# code to create new colum for chronically sick, use np.where 
df['chron_sick'] = np.where(df['PHYSHLTH']>=20,1,0)
##we define chronically sick, just give a reason
#- Those who are chronically sick are sick equal to or more than 15 days a month
##std above mean or something like that, boxplot, or histogram


```

## 5. Create a graphic that compares the percentage of people labeled as 'sick' in each state. 

```python
# code to calculate averages per state
sick_mean = df.groupby('_STATE')['chron_sick'].mean()
sick_mean
```

```python
sick_value = list(sick_mean.values)
```

```python
# create graphic to compare chronically sick by state
fig,ax = plt.subplots()

ax.bar(states,sick_value,width=0.8,align='center',color='r')
ax.set_title('Chronically Sick by State Comparison')
ax.set_ylabel('Percent of Sick')
ax.set_xlabel('States')
ax.axis('tight')

plt.show()
```
